```python
# Import the modules
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
```

```python
# Generate synthetic clusters
x, y = make_blobs(
    centers=4,
    random_state=40,
    n_samples=2000
)
```

```python
# Visualizing both classes
plt.scatter(x[:,0], x[:,1], c=y)
```
